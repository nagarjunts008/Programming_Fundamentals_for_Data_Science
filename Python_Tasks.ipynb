## Task 1:
Consider the Iris dataset. The dataset is organised as a single Comma Separated Values (CSV) file. The file contains data about three species
of iris.<br/>
Read the dataset into a suitable data structure and print on the screen its first 10 rows.<br/>
By considering only the Petal width, obtain the following statistical information for each one
of the three species of iris from the dataset:
- Measures of Central Tendency (mean and median)
- Measures of Dispersion (standards deviation, range)
Provide a complete Python code and the resulting statistical values.

```python
#import required libraries
import pandas as pd
from pandas.plotting import parallel_coordinates
import statistics
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Read dataset and print the first 10 rows
df = pd.read_csv('Iris.csv')
print(df.head(10))
```

```python
#Measures of Central Tendency

#Mean of Petal Width
petalwidth = df["Petal width"]
petalwidth_mean = statistics.mean(petalwidth)
print("Mean is :", petalwidth_mean)

#Median of Petal Width
petalwidth_median = statistics.median(petalwidth)
print("Median is :", petalwidth_median)
```

```python
# Measures of Dispersion

#Standard deviation of Petal Width
petalwidth_stddev = statistics.stdev(petalwidth)
print("Standand Deviation is :", petalwidth_stddev)

#Range of Petal Width
print("Range is:",min(petalwidth),"to",max(petalwidth))
```

## Task 2:
Identify which one of the following types of charts would provide optimal visualisation for a
single attribute of a single species from the Iris dataset:
- Line Chart
- Bar Chart
- Box Plot Chart<br/>
Justify your choice with maximum 5 sentences. Choose one species and one
attribute from the Iris dataset and visualise its values by using the selected type of chart. In this case, the data will be one-dimensional. Provide the Python code and the
resulting visualisation.

##### Identification:
The box plot chart is the optimal visualisation for a single attribute of a single species from the Iris dataset.

##### Justification:

The box plot was chosen as an optimal visualisation for this purpose because it is a standardized way of distributing the data. It is easy to identify properties like minimum, first quartile, median, third quartile and maximum.  Moreover, it can project the outliers and their values. Hence, Box plot chart is chosen over Line chart and Bar chart.

```python
# Data Visualisation considering Versicolor as species and Sepal Length as the attribute

df_Versicolor =df.loc[df['Species'] == "Versicolor"]
fig = plt.figure(figsize =(10, 7))
plt.boxplot(df_Versicolor["Sepal length"])

plt.xlabel("Versicolor")
plt.ylabel("Sepal Length")
plt.show()
```
